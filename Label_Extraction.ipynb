```python
from paddleocr import PaddleOCR # main OCR dependencies
from matplotlib import pyplot as plt # plot images
import cv2 #opencv
import os
```

# **Instantiate Model and Detect**

```python
ocr_model = PaddleOCR(lang='en')
```

```python
img_path = os.path.join('.', 'drug3.jpg')
```

```python
result = ocr_model.predict(img_path)
```

```python
result
```

```python
for res in result:
    for text in res['rec_texts']:
        print(text)
```

# **Visualise Results**

```python
# Save annotated image
for res in result:
    res.save_to_img("annotated.png")

# Show in notebook
from PIL import Image
import matplotlib.pyplot as plt
plt.figure(figsize=(40,20))
plt.imshow(Image.open("annotated.png"))
plt.axis("off")
plt.show()
```
